```python
import random
import math

data = [['見','貝'],['土','士'],['眠','眼']]
level = 1
col = 5
row = 4
number_data=['A','B','C','D','E']

def start_message():
  print('違う漢字の番号（例:A1)を入力してください')

def section_message():
  print('レベル:'+str(level))


def view_question():
  question_number = random.randint(0,2)
  mistake_number = random.randint(0,col*row-1)
  print('テバック：mistake_number:'+ str(mistake_number))
  question = data[question_number]
  print(question)
  question_str1 = '/ | '
  question_str2 = 'ー　ー　'
  x = 0
  while x < col:
    question_str1 += number_data[x] +' '
    x += 1
  question_str2 += 'ー　'*col
  print(question_str1)
  print(question_str2)
  x = 0
  y = 0
  while x < row:
    question_str = str(x+1) + '| '
    while y < col:
      if x*col + y == mistake_number:
        question_str += question[1]
      else:
        question_str += question[0]
      y += 1
    print(question_str)
    x += 1
    y = 0
  return mistake_number

def change_input_number(input_str):
  str_data = {'A':0,'B':1,'C':2,'D':3,'E':4}
  input_str_split = list(input_str)
  col_number = str_data[input_str_split[0]]
  row_number = int(input_str_split[1])-1
  input_number = col_number + row_number*col
  return input_number
  
def is_correct_number(mistake_number,input_number):
  if mistake_number == input_number:
    return True
  else:
    return False

def change_string(number):
  row_number = number_data[number % col]
  col_number = math.floor(number / row)+1
  return_number_str = row_number + str(col_number)
  return return_number_str


def view_result(is_correct,return_number_str):
  if is_correct:
    print('正解！')
  else:
    print('不正解')
    print('正解は'+return_number_str)

def play():
  section_message()
  mistake_number = view_question()
  choice = input('例:A1')
  print('テバック:choice = '+choice)
  input_number = change_input_number(choice)
  print('テバック:input_number:'+str(input_number))
  is_correct = is_correct_number(mistake_number,input_number)
  return_number_str = change_string(mistake_number)
  view_result(is_correct,return_number_str)

start_message()
play()
```
